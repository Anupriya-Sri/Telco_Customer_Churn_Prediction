```python
telco_cat = telco_churn_df.select_dtypes(include = 'object').copy().drop(columns=['customerID', 'Churn', 'TotalCharges'])

catfeatures_stats = pd.DataFrame(columns = ['column', 'values', 'unique_values', 'missing'])

tmp = pd.DataFrame()
#print(telco_cat.columns)

for c in telco_cat.columns:
    #print(c)
    tmp['column'] = [c]
    tmp['values'] = [telco_cat[c].unique()]
    tmp['unique_values'] = int(telco_cat[c].nunique())
    tmp['missing'] = telco_cat[c].isnull().sum()
    catfeatures_stats = catfeatures_stats.append(tmp)
    
catfeatures_stats
```
